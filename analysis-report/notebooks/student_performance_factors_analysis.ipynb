## Data Cleaning and Preparing for analysis

Data Checks to perform:

- Check Missing values
- Check Duplicates
- Check the number of unique values of each column
- Check various categories present in the different categorical column

```python
dataset.isna().sum()
```

```python
print(dataset["Teacher_Quality"].unique())
print(dataset["Parental_Education_Level"].unique())
print(dataset["Distance_from_Home"].unique())

print(dataset["Teacher_Quality"].mode())
print(dataset["Parental_Education_Level"].mode())
print(dataset["Distance_from_Home"].mode())

# Handling null values - categorical replaces with mode
dataset['Teacher_Quality'] = dataset['Teacher_Quality'].fillna(dataset['Teacher_Quality'].mode()[0])
dataset['Parental_Education_Level'] = dataset['Parental_Education_Level'].fillna(dataset['Parental_Education_Level'].mode()[0])
dataset['Distance_from_Home'] = dataset['Distance_from_Home'].fillna(dataset['Distance_from_Home'].mode()[0])

dataset.isna().sum()
```

```python
# Check for duplicates

dataset.duplicated().sum()
```

```python
# Checking the number of unique values of each column
dataset.nunique()
```

```python
# define numerical & categorical columns
numeric_features = [feature for feature in dataset.columns if dataset[feature].dtype != 'O']
categorical_features = [feature for feature in dataset.columns if dataset[feature].dtype == 'O']

print("Numeric features")
print(numeric_features)
print("Categorical features")
print(categorical_features)
```

```python
for item in categorical_features:
    print(item, ' ',dataset[item].unique())
```
